**Extracting images from PDF and storing them in a list**

```python
import os


#path of PDF containing Images
image_pdf_path = '/content/drive/MyDrive/Colab Notebooks/IEEE_Conference_Template__2_ (1).pdf'
#path of folder to put all the extracted images
output_folder = '/content/drive/MyDrive/Colab Notebooks/Output_Images'
extract_images_from_pdf(image_pdf_path, output_folder)

def get_image_paths(output_folder, extensions=['.jpg', '.jpeg', '.png', '.gif']):
    image_paths = []
    # Iterate through all files in the folder
    for root, dirs, files in os.walk(output_folder):
        for file in files:
            # Check if the file has one of the specified image extensions
            if any(file.lower().endswith(ext) for ext in extensions):
                # Construct the full path of the image file
                image_path = os.path.join(root, file)
                # Add the image path to the list
                image_paths.append(image_path)
    return image_paths

# Get the paths of all images in the folder
image_paths = get_image_paths(output_folder)

# Print the paths of all images
for image_path in image_paths:
    print(image_path)
```
